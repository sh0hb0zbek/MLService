# DataCleaning

```python
print('[DataCleaning.delete_unique()]')
try:
    df_delete_unique = data_cleaner.delete_unique(df)
    df_delete_unique
except Exception as e:
    print(e)
# there is no any unique values so it shows error
```

```python
print('[DataCleaning.delete_duplicate()]')
df_delete_duplicate = data_cleaner.delete_duplicate(df)
print('before', data_explore.shape(df))
print('after ', data_explore.shape(df_delete_duplicate))
# there is no any duplicated rows so nothing was deleted
df_delete_duplicate
```

```python
print('[DataCleaning.outlier()]')
df_outlier = data_cleaner.outlier(df, z=2)
print('before', data_explore.shape(df))
print('after ', data_explore.shape(df_outlier))
df_outlier
```
